# LinkedIn Connections

```python
def connection_list() :
    driver.get('https://www.linkedin.com/mynetwork/invite-connect/connections/')
    sleep(2)
    name = driver.find_elements(By.CSS_SELECTOR, 'span.mn-connection-card__name')
    link = driver.find_elements(By.CSS_SELECTOR, 'a.mn-connection-card__link')
    tag = driver.find_elements(By.CSS_SELECTOR, 'span.mn-connection-card__occupation')
    name_list = {}
    for i in range(len(name)):
        name_list[name[i].get_attribute('innerText')] = {
            'link' : link[i].get_attribute('href'),
            'tagline' : tag[i].get_attribute('innerText')
        }
    connections = pd.DataFrame(name_list)
    connections = connections.transpose()
    return connections
```

```python
p = 1
people = {}
company = input('enter a company: ')
flag = 1

while(flag == 1):
    url = f'https://www.linkedin.com/search/results/people/?keywords={company}&origin=CLUSTER_EXPANSION&page={p}&sid=%3Bs7'
    driver.get(url)
    sleep(3)
    driver.execute_script("window.scrollBy(0,1000)","")
    
    div = driver.find_elements(By.CSS_SELECTOR, 'div.entity-result__item')
    sleep(5)
    
    if(len(div)>0): 
        
        for i in range(len(div)):
            url = div[i].find_element(By.CSS_SELECTOR, 'a.app-aware-link').get_attribute('href')
            try:
                name = div[i].find_element(By.CSS_SELECTOR, 'a.app-aware-link>span>span[aria-hidden = "true"]').get_attribute('innerText')
            except:
                pass
            tag = div[i].find_element(By.CSS_SELECTOR, 'div.entity-result__primary-subtitle').get_attribute('innerText')
        
            people[name] = {
                'url' : url,
                'tagline' : tag
            }
    else:
        print('scraping completed!')
        flag = 0
        
    p += 1
```

```python
def click_next_button():
    try:
        next_btn = driver.find_element(By.CSS_SELECTOR, 'button[aria-label="Next"]')
        next_btn.click()
        return 0
    except:
        print('end of pages')
        return 1
```

```python
def company_people():
    flag = 0
    company = input('enter a company: ')
    url = f'https://www.linkedin.com/search/results/people/?keywords={company}&origin=CLUSTER_EXPANSION&sid=%3Bs7'
    driver.get(url)
    sleep(2)
    
    people = {}
    while(flag != 1):
        div = driver.find_elements(By.CSS_SELECTOR, 'div.entity-result__item')
        for i in range(len(div)):
            url = div[i].find_element(By.CSS_SELECTOR, 'a.app-aware-link').get_attribute('href')
            name = div[i].find_element(By.CSS_SELECTOR,'span[dir="ltr"]').find_element(By.CSS_SELECTOR, 'span').get_attribute('innerText')
            tag = div[i].find_element(By.CSS_SELECTOR, 'div.entity-result__primary-subtitle').get_attribute('innerText')
        
            people[tag[:5]] = {
                'url' : url,
                'tagline' : tag
            }
        sleep(2)
        driver.execute_script("window.scrollBy(0,1000)","")
        sleep(2)
        flag = click_next_button()
            
    people_df = pd.DataFrame(people)
    people_df = people_df.transpose()
    return people_df
```
